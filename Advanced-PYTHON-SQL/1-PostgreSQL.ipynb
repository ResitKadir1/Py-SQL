# 5 -ROLLUP


```python
#sum up all values with in group
cursor.execute("""
select category_id,
	product_name,
	count(*),
	min(price) ,
	max(price),
	avg(price)
from inventory.products

group by rollup (category_id, product_name)

order by category_id, product_name;


""")
result = cursor.fetchall();
for i in result:
    print(i)
```

# 6--CUBE 

There is only one major difference between the functionality of the ROLLUP operator and the CUBE operator. ROLLUP operator generates aggregated results for the selected columns in a hierarchical way. On the other hand, CUBE generates a aggregated result that contains all the possible combinations for the selected columns.

GROUPING SETS is a controllable and scalable option, so I prefer to using it in lieu of ROLLUP and CUBE.

```python
cursor.execute("""
select category_id,
	size,
	count(*),
	min(price) as "lowest price",
	max(price) as "highest price",
	avg(price) as "average price"
from inventory.products
group by cube (category_id, size)
order by category_id, size;


""")
result = cursor.fetchall();
for i in result:
    print(i)
```

```python
#-small producs--> count all filter to whichs size smaller than 16 named it as count small
#get avarage of it whichs smaller than 16 and name it as avarage price small

#--> output--> 
    #Count small--> 35
    #avarage price small-> 12.4
    
cursor.execute("""

select 
    
    category_id,
	count(*) as "count all",
	avg(price) as "average price",
	
    -- small products
	count(*) filter (where size <=16) as "count small",
	avg(price) filter (where size <= 16) as "average price small",
    
	-- large products
	count(*) filter (where size >16) as "count large",
	avg(price) filter (where size >16) as "average price large"
    
from inventory.products
group by rollup (category_id)
order by category_id DESC

""")
result = cursor.fetchall();
for i in result:
    print(i)
```

```python
# 2
cursor.execute("""
select customer_id,
	count(*) filter (where order_date >= '2021-03-01' and order_date <= '2021-03-31') as "March",
	count(*) filter (where order_date between '2021-04-01' and '2021-04-30') as "April"

from sales.orders
group by customer_id;
""")
result = cursor.fetchall();
for i in result:
    print(i)
```

*3-From sales.order_lines,find the total qauntity of each produc sold;use the SUM()aggregate function to add up the product quantities

```python
# 3
cursor.execute("""
-- Quantity of each product sol

select * from sales.order_lines;

select sku, sum(quantity) as "Total Sold"
from sales.order_lines
group by rollup (sku)
order by sum(quantity) DESC;
""")
result = cursor.fetchall();
for i in result:
    print(i)
```
